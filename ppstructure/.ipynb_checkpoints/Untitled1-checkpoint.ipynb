```python
from tqdm import tqdm
```

```python
def add_question_to_answer_df(img_path ,answer_df):
    assert os.path.exists(img_path)
    assert hasattr(answer_df, "size")

    img = Image.open(img_path)
    img_array = np.asarray(img)
    if len(img_array.shape) == 3:
        #img_array = img_array[:, :, :3]
        pass
    else:
        assert len(img_array.shape) == 2
        h, w = img_array.shape
        img_array = img_array.reshape([h, w, 1])
        img_array = np.concatenate([img_array] * 3, axis = -1)
    assert len(img_array.shape) == 3
    img_array = img_array[:, :, :3]

    req = []
    for i, r in tqdm(answer_df.iterrows()):
        d = r.to_dict()
        answer = d["transcription"]
        qa_d = demo_process_vqa(img_array, answer)
        pred_question = qa_d["answer"]
        d["question"] = pred_question
        req.append(d)
    return pd.DataFrame(req).rename(columns = {"transcription": "answer"})
```
